```python
from sklearn.model_selection import train_test_split
```

```python
x = df[["Time on App", "Time on Website","Avg. Session Length", "Length of Membership"]]
y = df["Yearly Amount Spent"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Training model
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
```

```python
ln.fit(X_train, y_train)
```

```python
ln.coef_
```

```python
ln.intercept_
```

```python
cdf = pd.DataFrame(ln.coef_, x.columns, columns = ["Coef"])
print(cdf)
```

```python
# Predictions
pred = ln.predict(X_test)
pred
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.scatterplot(pred, y_test)
plt.xlabel("Prediction")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
print("MSE :" ,mean_squared_error(y_test,pred)) #  
print("MAE :" ,mean_absolute_error(y_test,pred))# Average distance of a everypoint to actual line.
print("RMSE :",math.sqrt(mean_squared_error(y_test,pred)))
```

```python
residuals = y_test-pred
residuals
```

```python
sns.displot(residuals, bins = 16, kde = True) # Test Normality
```

```python
import pylab
import scipy.stats as stats

stats.probplot(residuals, dist = "norm", plot = pylab)
pylab.show()
```
